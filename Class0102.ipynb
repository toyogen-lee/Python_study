```python
#반복문의 3요소 - 값(초기), 조건(종료), 스텝(보폭)

dan = 8
i = 1  #값
while i<=3:  #조건
  print("{} * {} = {}".format(dan,i,dan*i))
  i+=1  #스텝
print(i)
```

```python
#다중 반복문

i = 1
while i <= 3:
  j = 1
  while j <= 3:
    print(j)
    j+=1
  i+=1
```

```python
# 문제 - 1부터 3까지 10번 반복해서 출력해주세요.
i = 1
while i <= 10:
  j = 1
  while j <= 3:
    print(j)
    j+=1
  i+=1
```

```python
# 문제 - 구구단을 출력해주세요.(2단 ~ 9단까지)
i = 2
while i<=9:
  j=1
  while j<=9:
    print("{} * {} = {}".format(i,j,i*j))
    j+=1
  i+=1
```

```python
# 문제 - 구구구단을 출력해주세요.
a = 1
while a <= 9:
  b = 1
  while b <= 9:
    c=1
    while c <= 9:
      print("{} * {} * {} = {}".format(a,b,c,a*b*c))
      c += 1
    b += 1
  a += 1
```
