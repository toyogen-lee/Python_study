```python
#elif | else : 나이에 따른 분류
age = 14
if age>=0 and age<10:
  print("유아/아동")
elif age>=10 and age<20:
  print("10대")
elif age>=20 and age<30:
  print("20대")
elif age>=30 and age<40:
  print("30대")
else:               #elif age>=40: 불필요한 코드 생략
  print("40대 이상")
#if+else : 이지선다형 코드else:               #elif age>=40: 불필요한 코드 생략
  print("40대 이상")
  #if+else : 이지선다형 코드
```

```python
# 문제 : 구구단 8단을 출력해주세요.
# 조건 : for, while문을 사용할 수 없습니다.

print("v1")
print("8 * 1 = 8")
print("8 * 2 = 16")
print("8 * 3 = 24")
# 8 * 9까지 같은 방식

print("v2")
s = "8 * 1 = 8"
s = s + "\n8 * 2 = 16"
s += "\n8 * 3 = 24"
print(s)
# 8 * 9까지 같은 방식

print("v3")
print("{} * {} = {}".format(8, 1, 8))
print("{} * {} = {}".format(8, 2, 16))
print("{} * {} = {}".format(8, 3, 24))
# 8 * 9까지 같은 방식

print("v4")
i = 1
print("{} * {} = {}".format(8, i, 8))
i = i + 1
print("{} * {} = {}".format(8, i, 16))
i += 1
print("{} * {} = {}".format(8, i, 24))
# 8 * 9까지 같은 방식

print("v5")
dan = 8
i = 1
print("{} * {} = {}".format(dan, i, dan * i))
i = i + 1
print("{} * {} = {}".format(dan, i, dan * i))
i += 1
print("{} * {} = {}".format(dan, i, dan * i))
# 8 * 9까지 같은 방식
```
